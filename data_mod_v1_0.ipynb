### Search operation for institutions' urls

```python
# Function to fetch the college URL using Google Custom Search API
def get_college_url(college_names, service, cse_id, delay=2):
    try:
        if pd.isna(college_names):
            return ""
        print(f"Fetching URL for: {college_names}")
        res = service.cse().list(
            q=college_names,
            cx=cse_id,
        ).execute()
        print(f"Response received for: {college_names}")

        search_items = res.get('items', [])
        if search_items:
            return search_items[0].get('link')
        else:
            return ""
    except Exception as e:
        print(f"Error fetching URL for {college_names}: {e}")
        return ""
    finally:
        time.sleep(delay)

# Build the service
service = build("customsearch", "v1", developerKey=API_KEY)
```

### Search operation for wayback machine urls & snapshots


```python
# Function to save URLs to the Wayback Machine with error handling and exponential backoff
def get_existing_wayback_url(url, retries=5, delay=10):
    """
    Retrieves the URL of the closest existing snapshot from the Wayback Machine for a given URL.

    Args:
    - url (str): The URL to check for existing snapshots.
    - retries (int): Number of times to retry in case of failure (default is 5).
    - delay (int): Delay in seconds between retries (default is 10).

    Returns:
    - str: The URL of the closest existing snapshot, "No snapshot available" if none found,
           or "Error retrieving snapshot" if an error occurs after all retries.
    """
    if not url:
        return ""

    api_url = f"http://archive.org/wayback/available?url={url}"
    for attempt in range(retries):
        try:
            response = requests.get(api_url)
            if response.status_code == 200:
                data = response.json()
                snapshots = data.get("archived_snapshots", {})
                if "closest" in snapshots:
                    return snapshots["closest"]["url"]
                else:
                    return "No snapshot available"
            else:
                print(f"Failed to retrieve snapshot for {url}. Status code: {response.status_code}")
        except requests.RequestException as e:
            print(f"Request exception for {url}: {e}")

        print(f"Retrying in {delay} seconds...")
        time.sleep(delay)

    return "Error retrieving snapshot"
```

### Adding Google Scholar profiles

```python
# Setup logging
logging.basicConfig(filename='scholar_scraper.log', level=logging.INFO)

# Initialize user agent
ua = UserAgent()

def get_scholar_profile_url(name, delay=2):
    """
    Retrieves the Google Scholar profile URL for a given name using Google Custom Search API.

    Args:
    - name (str): The name of the scholar to search for.

    Returns:
    - str: The URL of the scholar's profile or "not found" if no profile is found.
    """
    try:
        query = f"{name} site:scholar.google.com"
        url = f"https://www.googleapis.com/customsearch/v1?q={query}&cx={CSE_ID}&key={API_KEY}"
        response = requests.get(url)

        if response.status_code == 200:
            results = response.json()
            if 'items' in results:
                return results['items'][0]['link']
            else:
                return "not found"
        elif response.status_code == 429:
            logging.warning(f"Rate limit hit for {name}. Waiting for 60 seconds before retrying...")
            time.sleep(60)  # Wait for 60 seconds before retrying
            return get_scholar_profile_url(name, delay)
        else:
            logging.error(f"Error fetching Scholar profile URL for {name}: {response.status_code} {response.text}")
            return "not found"
    except Exception as e:
        logging.error(f"Error fetching Scholar profile URL for {name}: {e}")
        return "not found"
```

### Scrapping function definition

```python
# Define required columns
required_columns = ['names_scholar', 'other_names', 'position', 'field', 'total_citation', 'h-index', 'i10-index', 'profile_since', 'co-authors']

# Initialize AsyncHTMLSession
asession = AsyncHTMLSession()



# Function to scrape data from Google Scholar
async def get_scholar_data(scholar_profile_url):
    scholar_data = {key: 'not found' for key in required_columns}

    if pd.isna(scholar_profile_url) or scholar_profile_url.strip() == '' or scholar_profile_url == 'not found':
        return scholar_data

    try:
        response = await asession.get(scholar_profile_url)
        await response.html.arender(timeout=20)  # Render JavaScript with a timeout of 20 seconds

        name_element = response.html.find('#gsc_prf_in', first=True)
        if name_element:
            scholar_data['names_scholar'] = name_element.text

        if response.html.find('#gsc_prf_ion_lbl', first=True):
            other_names_element = response.html.find('#gs_prf_ion_txt', first=True)
            if other_names_element:
                scholar_data['other_names'] = other_names_element.text

        position_element = response.html.find('div#gsc_prf_i > div:nth-child(2)', first=True)
        if position_element:
            scholar_data['position'] = position_element.text

        fields_elements = response.html.find('div#gsc_prf_int > a')
        if fields_elements:
            scholar_data['field'] = ', '.join([field.text for field in fields_elements])

        total_citation_element = response.html.find('.gsc_rsb_std', first=True)
        if total_citation_element:
            scholar_data['total_citation'] = total_citation_element.text

        h_index_element = response.html.find('table#gsc_rsb_st > tbody > tr:nth-child(2) > td:nth-child(2)', first=True)
        if h_index_element:
            scholar_data['h-index'] = h_index_element.text

        i10_index_element = response.html.find('table#gsc_rsb_st > tbody > tr:nth-child(3) > td:nth-child(2)', first=True)
        if i10_index_element:
            scholar_data['i10-index'] = i10_index_element.text

        profile_since_element = response.html.find('div[role="navigation"] th:nth-child(3)', first=True)
        if profile_since_element:
            scholar_data['profile_since'] = profile_since_element.text

        # Fetch co-authors
        co_authors_elements = response.html.find('#gsc_rsb_co')
        if co_authors_elements:
            co_authors_text = co_authors_elements[0].text
            co_authors = [line.split('\n')[0] for line in co_authors_text.split('\n') if line.strip() and not line.startswith('#')]
            if co_authors and co_authors[0].lower().startswith('co-authors'):
                co_authors = co_authors[1:]
            if co_authors:
                scholar_data['co-authors'] = ', '.join(co_authors)

    except Exception as e:
        logging.error(f"Rendering timeout or error for URL: {scholar_profile_url}: {e}")

    return scholar_data
```

### Scrapping action




```python
# Function to process a batch
async def process_batch(df_batch):
    # Ensure all required columns are initialized with dtype 'object'
    for col in required_columns:
        if col not in df_batch.columns:
            df_batch[col] = pd.Series(dtype='object')

    tasks = [get_scholar_data(row['scholar_profile_url']) for _, row in df_batch.iterrows()]
    results = await asyncio.gather(*tasks)

    for index, scholar_data in enumerate(results):
        profile_url = df_batch.iloc[index]['scholar_profile_url']  # Get profile URL
        for key, value in scholar_data.items():
            # Ensure that we assign to the correct column
            df_batch.at[df_batch.index[index], key] = value
        name = scholar_data.get('names_scholar', 'Unknown Name')  # Get name or default to 'Unknown Name'
        print(f"Processed {name} - {profile_url}")
        logging.info(f"Processed {name} - {profile_url}")

# Function to process profiles in batches
async def process_profiles_in_batches(df, batch_size=10, delay=30):
    # Initialize DataFrame with required columns if they don't exist
    for col in required_columns:
        if col not in df.columns:
            df[col] = pd.Series(dtype='object')

    total_rows = df.shape[0]
    for start in range(0, total_rows, batch_size):
        end = min(start + batch_size, total_rows)
        df_batch = df.iloc[start:end]
        await process_batch(df_batch)
        print(f"Processed rows {start} to {end - 1}")
        logging.info(f"Processed rows {start} to {end - 1}")
        if end < total_rows:
            print(f"Waiting for {delay} seconds before processing the next batch...")
            logging.info(f"Waiting for {delay} seconds before processing the next batch...")
            await asyncio.sleep(delay)  # Pause between batches

# Run the async function within an existing event loop
loop = asyncio.get_event_loop()
loop.run_until_complete(process_profiles_in_batches(df))

print("Processing completed. Scholar profile data saved to the CSV file.")
logging.info("Processing completed. Scholar profile data saved to the CSV file.")

# Specify the columns you want to view
columns_to_view = required_columns

# Display the first 50 rows of the specified columns
print(df[columns_to_view].head(50))
```
